## Heatmap for 2018-19 Influenza Season Week 4 ending Jan 28, 2019

### Import usmap and ggplot2 for plotting the US map

```r
library(usmap)
library(ggplot2)
```

### Read the data for week 4 from csv and use ACTIVITY.LEVEL column for more fine grained display of ILI Activity Levels

```r
states_df <- read.csv("./FluViewPhase2Data/StateDatabyWeekforMap_2018-19week4.csv")

colnames(states_df)[1] = "state"
# strip "Level " from the column values
states_df$ACTIVITY.LEVEL = as.numeric(gsub('Level ', '', states_df$ACTIVITY.LEVEL))
```

### Use the ACTIVITY.LEVEL column to display heatmap 

```r
plot_usmap(data = states_df, values = "ACTIVITY.LEVEL", lines = "black", regions = c("county")) +
  ggtitle("2018-19 Influenza Season Week 4 ending Jan 28, 2019") +
  scale_fill_continuous(
    name = "ILI Activity Level",
    type="gradient",
    low = "green", high = "red",
    breaks = seq(1, 10),
    labels = c(rep("Minimal", 3), rep("Low", 2), rep("Moderate", 2), rep("High", 3)),
    na.value = "transparent"
  ) +
  guides(colour = "colorbar") +
  theme(legend.position = "right", legend.key.height = unit(2, "cm"), plot.title = element_text(hjust = 0.5))
```

## Analysis:
- The above plot implies that southern and south-east states have more Influenza activity during the week 4 of 2019 (28th January 2019)
- It is interesting to see that Alaska has higher influenza activity reported by CDC, but the results of twitter analysis for flu showed that Alaska has lower activity which can be explained by the fact that Alaska might have lesser active users on twitter and even lesser people tweeting about it.
