# Problem 1:
- We use poison distribution to generate random data.
- then we plot line graph showing sales for two datasets in one plot.

```r
# Problem 1
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,34)  # random numbers, Poisson distribution, mean at 34, 12 numbers

par(bg="cornsilk")
plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" )
title(main="Sales by Month")


lines(sales2, type="o", pch=22, lty=2, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
```

# Problem 2
- We read data from the chosen file as a table.
- We generate a barplot to visualize two sales vectors side by side.
- We can infer from the data that the products in comparison are complementary goods i.e. when sales of one product is more the sales of other product is less and vice-versa.

```r
# Problem 2
sales<-read.table(file.choose(), header=T)
sales  # to verify that data has been read
barplot(as.matrix(sales), main="Sales Data", ylab= "Total",beside=T, col=rainbow(5))
```

# Problem 3
- Here we draw a boxplot to visualize the overall spread of data so as to get a better overall spread of the data.
- From the boxplot below we can infer that sales of product 2 is better overall than that of product 1 which can be explained by lower median, minimum and maximum values of sales of product 1 against product 2.

```r
# Problem 3
fn<-boxplot(sales,col=c("orange","green"))$stats

text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
```

# Problem 4:
- We overlay data of facebook and apple stock price to compare historical data.
- According to the given data we can infer that Facebook stock price has pretty much always been greater than Apple's stock price during the specified period.

```r
# Problem 4
fb1<-read.csv(file.choose())
aapl1<-read.csv(file.choose())
par(bg="cornsilk")
plot(aapl1$Adj.Close, col="blue", type="o", ylim=c(120,220), xlab="Days", ylab="Price" )
lines(fb1$Adj.Close, type="o", pch=22, lty=2, col="red")
legend("topright", inset=.05, c("Apple","Facebook"), fill=c("blue","red"), horiz=TRUE)
hist(aapl1$Adj.Close, col=rainbow(8))
```

# Problem 6:
I have used an API key to register with google maps and have provided cities vector to te geocode function to get the latitude and longitude corresponding to those cities and have plotted some of the cities using $ sign

```r
# Problem 6
# on windows use command > set MAPS_KEY=apiKey before running the notebook
maps_key <- Sys.getenv("MAPS_KEY")
library("ggmap")
library("maptools")
library("maps")
register_google(key = maps_key) 
visited <- c("SFO", "Chennai", "London", "Melbourne", "Lima,Peru", "Johannesbury, SA")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
map("world", fill=TRUE, col="white", bg="lightblue", ylim=c(-60, 90), mar=c(0,0,0,0))
points(visit.x,visit.y, col="red", pch=36)
```

We perform the same procedure to plot cities in the united states 

```r
# Problem 6
library("ggmap")
library("maptools")
library(maps)
visited <- c("SFO", "New York", "Buffalo", "Dallas, TX")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
map("state", fill=TRUE, col=rainbow(50), bg="lightblue", mar=c(0,0,0,0))
points(visit.x,visit.y, col="yellow", pch=36)
```
